```python
def remove_duplicate_columns(df, column_name):
    cols = [col for col in df.columns if col == column_name]
    if len(cols) > 1:
        df = df.loc[:, ~df.columns.duplicated(keep='first')]
    return df

estados = remove_duplicate_columns(estados, 'estadium')

print(estados.columns)
```

```python
def filter_by_city(estados, estadium):
    return estados[estados['estadium'] == estadium]

def find_similar_restaurants(name_y, estadium, num_recommendations=5):
    city_restaurants = filter_by_city(estados, estadium)
    
    if name_y not in city_restaurants['name_y'].values:
        raise ValueError(f"El restaurante '{name_y}' no existe cerca del estadio '{estadium}'.")

    city_restaurants = city_restaurants.drop_duplicates(subset=['name_y'])
    
    restaurant_matrix = city_restaurants.pivot_table(index='name_y', values=['avg_rating', 'num_of_reviews'])
    
    mlb = MultiLabelBinarizer()
    categories = mlb.fit_transform(city_restaurants['category'])
    categories_df = pd.DataFrame(categories, index=city_restaurants['name_y'], columns=mlb.classes_)
    
    combined_matrix = pd.concat([restaurant_matrix, categories_df], axis=1)
    
    combined_matrix_filled = combined_matrix.fillna(0)
    
    similarity_matrix = cosine_similarity(combined_matrix_filled)
    similarity_df = pd.DataFrame(similarity_matrix, index=combined_matrix.index, columns=combined_matrix.index)
    
    similar_restaurants = similarity_df[name_y].sort_values(ascending=False).head(num_recommendations + 1).index.tolist()
    similar_restaurants.remove(name_y)
    
    recommendations = estados[estados['name_y'].isin(similar_restaurants) & (estados['estadium'] == estadium)].drop_duplicates(subset=['name_y']).head(num_recommendations)
    
    return recommendations[['name_y', 'avg_rating', 'url', 'estadium']]

name_y = "Florence Pizza & Pasta"
estadium = "Geodis Park"
recommendations = find_similar_restaurants(name_y, estadium)
print(f'Recomendaciones para {name_y} cerca de {estadium}:')
print(recommendations)
```
